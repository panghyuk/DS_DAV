```python
# 필요한 라이브러리 불러오기
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px

# seaborn 색상 변경
sns.set_palette("pastel")
```

## '데이터 전처리'에서 저장했던 csv 파일 불러오기

```python
total = pd.read_csv("total.csv")
total
```

## 'Total' 데이터 프레임 데이터 타입 확인

```python
total.dtypes
```

## 변수 간 상관 관계 확인

```python
plt.figure(figsize=(10,8))
sns.heatmap(total.corr(),cmap="bwr",annot=True)
```

## 상관관계가 있는 변수끼리 묶어서 Heatmap 시각화

```python
sns.clustermap(total.corr(), annot=True, cmap="bwr")
```

## 변수끼리 관계 확인을 위한 'pairplot' 시각화

```python
sns.pairplot(total)
```

## 2018년 기준 GDP 상위 10개국, 하위 10개국 시각화

```python
top10 = total.set_index('Country Name')['2018_gdp'].nlargest(10).to_frame()
bottom10 = total.set_index('Country Name')['2018_gdp'].nsmallest(10).to_frame()

subdf = pd.concat([top10, bottom10], axis = 0)
subdf = subdf.sort_values(by = "2018_gdp")

fig = plt.figure(figsize = (15,10))

cmap = plt.cm.coolwarm_r
norm = plt.Normalize(vmin = subdf['2018_gdp'].min(), vmax = subdf['2018_gdp'].max())

colors = [cmap(norm(value)) for value in subdf['2018_gdp']]

g = plt.barh(y = subdf.index, width = subdf["2018_gdp"], color = colors)
plt.ylabel("Country Name", fontsize = 16)
plt.xlabel("GDP in 2018", fontsize = 16)
plt.title("Top-10 highest GDP and lowest GDP Countries", fontdict = {'size':20})

for rect in g.patches:
    x = rect.get_x()
    y = rect.get_y()
    height = rect.get_height()
    width = rect.get_width()
    
    plt.annotate(f"{width}", (width, y + height / 2), ha = 'left', va = 'center')

plt.show()
```

# 2018년 기준 Ladder Score(Happiness Score) 상위 10개국, 하위 10개국 시각화

```python
top20 = total.set_index('Country Name')['Ladder Score'].nlargest(20).to_frame()

subdf = top20.sort_values(by="Ladder Score",ascending=False)

fig = plt.figure(figsize = (15,10))

cmap = plt.cm.coolwarm_r
norm = plt.Normalize(vmin = subdf['Ladder Score'].min(), vmax = subdf['Ladder Score'].max())

colors = [cmap(norm(value)) for value in subdf['Ladder Score']]

g = plt.barh(y = subdf.index, width = subdf["Ladder Score"], color = colors)
plt.ylabel("Country Name", fontsize = 16)
plt.xlabel("Happiness Score", fontsize = 16)
plt.title("Top-20 most Happiest Countries", fontdict = {'size':20})

for rect in g.patches:
    x = rect.get_x()
    y = rect.get_y()
    height = rect.get_height()
    width = rect.get_width()
    
    plt.annotate(f"{width}", (width, y + height / 2), ha = 'left', va = 'center')

plt.show()
```

# 대륙 별, 나라 별 Ladder Score 시각화

```python
fig = px.sunburst(data_frame = total,
                  path = ["Regional indicator", "Country Name"],
                  values = "Ladder Score",
                  color = "Ladder Score",
                  width = 900, 
                  height = 900)
fig.show()
```

## 대륙 별 평균 Ladder Score(Happiness Score) 시각화

```python
g = total.groupby("Regional indicator")['Ladder Score'].mean().sort_values().plot.barh(figsize = (15,8), color = "orange")

g.set_xlabel("Happiness Score", fontsize = 16)
g.set_title("Average happiness score in each region", fontdict = {"size":20})

for rect in g.patches:
    x = rect.get_x()
    y = rect.get_y()
    height = rect.get_height()
    width = rect.get_width()
    
    plt.annotate(f"{width:.3f}", (width, y + height / 2), ha = 'left', va = 'center')
    
plt.show()
```

## 대륙 별 평균 GDP 시각화

```python
g = total.groupby("Regional indicator")['2018_gdp'].mean().sort_values().plot.barh(figsize = (15,8), color = "orange")

g.set_xlabel("2018 GDP", fontsize = 16)
g.set_title("Average GDP in each region", fontdict = {"size":20})

for rect in g.patches:
    x = rect.get_x()
    y = rect.get_y()
    height = rect.get_height()
    width = rect.get_width()
    
    plt.annotate(f"{width:.3f}", (width, y + height / 2), ha = 'left', va = 'center')
    
plt.show()
```
