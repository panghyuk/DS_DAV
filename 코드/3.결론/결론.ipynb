```python
# 필요한 라이브러리 불러오기
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
sns.set_palette("pastel")
```

## Total이라는 이름의 데이터 프레임 불러오기

```python
total = pd.read_csv("total.csv", index_col=0)
total
```

## well being 정도를 나타내는 데이터 불러오기
출처 : https://stats.oecd.org/index.aspx?DataSetCode=BLI

```python
well_being = pd.read_csv('current well-being.csv')
well_being
```

```python
well_being.isnull().sum()
```

#### 데이터가 너무 많고, null값이 다수 존재하므로 2010년 이후 데이터들의 평균값으로 해당 국가의 value값 대체
#### 필요없는 데이터는 삭제

```python
df = well_being[well_being['TIME']>=2010]
df.drop(df[df['TYPE_VAR']=='DEP'].index, axis=0, inplace=True)
df.drop(df[['LOCATION','TYPE_VAR','Type of indicator','WB','Current/Future Well-being','SEX',
            'Sex','AGE','Age','EDUCATION','Education','Time','Flag Codes','Flags']], axis=1, inplace=True)
df
```

#### 국가와 웰빙 지표 기준 groupby

```python
df_mean = df.groupby(['Country','Indicator']).mean()
df_mean
```

#### 국가를 x축, 웰빙 지표를 y축으로 데이터프레임 수정

```python
df_all = pd.DataFrame(index=df['Country'].unique(), columns=df['Indicator'].unique())

for i in df['Country'].unique():
    for j in df['Indicator'].unique():

        try:
            df_all.loc[i][j] = df_mean.loc[i,j]['Value']
        except:
            pass
df_all
```

#### 각 지표마다 범위가 다르므로, 모든 데이터들에 대해 정규화를 진행 (0과 1 사이로 정규화)

```python
df_scale = df_all.copy()
for i in df_all.columns:
    arr = np.array(df_all[i]).reshape(-1,1)
    scale = MinMaxScaler().fit_transform(arr)
    df_scale[i] = scale
df_scale
```

#### total의 gdp, co2 열과 scale 데이터프레임 합치기

```python
df_total = pd.merge(df_scale,total[['2018_gdp','2018_co2']], left_index=True, right_index=True, how='inner')
df_total
```

#### 필요없는 지표는 제거
#### 결측치는 평균값으로 대체

```python
df_total.drop(df_total[['Adult skills (literacy)','Adult skills (numeracy)','Gender wage gap','Gender gap in working hours',
                        'Road deaths', 'Student skills (reading)','Student skills (maths)','Student skills (science)','Time off']],
                          axis=1, inplace=True)

df_total.fillna(df_total.mean(), inplace=True)
df_total
```

```python
df_total.columns
```

### 영국의 신경제재단이 도입한 행복지수 계산법 =>    행복지수 = 웰빙 * 기대수명 * 부의평등/생태발자국

#### 웰빙은 위 각 지표에서 긍정적인 값은 더하고 부정적인 값은 뺐다.
##### - 긍정적인 값 : 가정 수입, 가정 부유 정도, 고용률, 인터넷 가능한 가정, 건강 인지 정도, 사회적 지지도, 정부로의 발언권, 투표율, 야간 안전 정도, 삶의 만족도, 시간 활용의 만족도, 녹지대로의 접근성, 사회적 상호작용에 소모하는 시간, 주택구매 용이성, 수입
##### - 부정적인 값 : 노동 시장의 불안성, 자살/알코올/약물로 인한 죽음, 살인, 부정적 영향의 균형, 우울증, 장기간 비급여 노동 시간

#### 기대수명 : Life expectancy at birth
#### 부의평등 : gdp
#### 생태발자국 : co2

```python
happy = []
for i in df_total.index:
    idx = df_total.loc[i]
    wellbeing = (idx['Household income']+idx['Household wealth']+idx['Employment rate']+idx['Households with internet access at home ']+
                idx['Perceived health']+idx['Social support']+idx['Having a say in government']+idx['Voter turnout ']+
                idx['Feeling safe at night']+idx['Life satisfaction']+idx['Satisfaction with time use']+idx['Access to green space']+
                idx['Time spent in social interactions']+idx['Satisfaction with personal relationships']+idx['Housing affordability']+
                idx['Earnings']
                -idx['Labour market insecurity']-idx['Deaths from suicide, alcohol, drugs']-idx['Homicides']
                -idx['Negative affect balance']-idx['Self-reported depression']-idx['Long unpaid working hours'])
    
    life_exp = idx['Life expectancy at birth']
    wealth = idx['2018_gdp']
    eco = idx['2018_co2']
    
    happy_index = wellbeing * life_exp * wealth / eco
    happy.append(happy_index)
df_total['Happy Planet Index'] = happy
df_total
```

```python
result = df_total.sort_values(by=['Happy Planet Index'],axis=0, ascending=False)
result
```

### 스위스, 스웨덴, 노르웨이, 덴마크, 아일랜드 등..

```python
fig = plt.figure(figsize=(15,10))

cmap = plt.cm.coolwarm_r
norm = plt.Normalize(vmin=result['Happy Planet Index'][:20].min(), vmax=result['Happy Planet Index'][:15].max())

colors = [cmap(norm(value)) for value in result['Happy Planet Index'][:20]]

g = plt.barh(y=result.iloc[:20].index, width=result['Happy Planet Index'][:20], color=colors)
plt.ylabel("Country Name", fontsize=16)
plt.xlabel("Happiness Score", fontsize=16)
plt.title("Top-20 most Happiest Countries", fontdict={'size':20})

for rect in g.patches:
    x = rect.get_x()
    y = rect.get_y()
    height = rect.get_height()
    width = rect.get_width()
    
    plt.annotate(f"{width}", (width, y+height/2), ha='left', va='center')

plt.show()
```
