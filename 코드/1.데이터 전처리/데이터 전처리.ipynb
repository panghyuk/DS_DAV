```python
# 필요한 라이브러리 불러오기
import numpy as np
import pandas as pd 
```

# 데이터 불러와서 데이터 프레임에 저장

- [World Happines Data](https://worldhappiness.report)<br>
- [GDP(current US$)](https://data.worldbank.org/indicator/NY.GDP.MKTP.CD)<br>
- [GDP growth(Annual %)](https://data.worldbank.org/indicator/NY.GDP.MKTP.KD.ZG)<br>
- [CO2 emissions](https://data.worldbank.org/indicator/EN.ATM.CO2E.KT)<br>
- [Life expectancy at birth,total(years)](https://data.worldbank.org/indicator/SP.DYN.LE00.IN)<br>

### World Happiness Data

```python
happy = pd.read_csv("world_happiness.csv")
happy
```

### GDP Data

```python
gdp = pd.read_csv("gdp.csv")
gdp
```

```python
gdp = gdp[["Country Name","2018_gdp"]]
gdp
```

### GDP growth Data

```python
gdp_growth = pd.read_csv("gdp_growth.csv")
gdp_growth
```

```python
gdp_growth = gdp_growth[['Country Name','2018_growth']]
gdp_growth
```

### Life Expectancy Data

```python
life = pd.read_csv("life expectancy.csv")
life
```

```python
life = life[["Country Name","2018_life"]]
life
```

### CO2 Data

```python
co2 = pd.read_csv('co2_emissions.csv')
co2
```

```python
co2 = co2[["Country Name","2018_co2"]]
co2
```

# 'Total'이라는 이름의 DataFrame 생성하기

### 'merge'를 활용하여 전체 데이터 프레임 구성

```python
total = pd.merge(happy, gdp,how='inner',on = 'Country Name')
total
```

```python
total = pd.merge(total,gdp_growth,how = 'inner',on = 'Country Name')
total
```

```python
total = pd.merge(total,life,how = 'inner',on = 'Country Name')
total
```

```python
total = pd.merge(total,co2,how = 'inner',on = 'Country Name')
total
```

## 2021년 World Happiness 데이터를 참고하여 대륙 칼럼 추가

```python
happy_21 = pd.read_csv("world_happiness_21.csv")
happy_21 = happy_21[["Country Name","Regional indicator"]]
happy_21
```

```python
total = pd.merge(total,happy_21, how = "inner", on = "Country Name")
total
```

# 완성된 DataFrame .csv 파일로 만들기

```python
total.to_csv("total.csv",index=False)
```
